# Simple Tower Defense Simulation
Run the code cell below to play a basic tower-defense simulation.

```python
import random

money = 100
lives = 10
wave = 1
towers = 1

print('Tower Defense Simulation Started!')

while lives > 0 and wave <= 5:
    enemies = 3 + wave
    print(f'\nWave {wave} — Enemies: {enemies}')

    for e in range(enemies):
        enemy_hp = 10 + wave * 2
        damage = towers * random.randint(4, 8)

        if damage >= enemy_hp:
            money += 10
            print('Enemy defeated! +10 money')
        else:
            lives -= 1
            print('Enemy escaped! -1 life')

    if money >= 50:
        towers += 1
        money -= 50
        print('Built a new tower!')

    wave += 1

print('\nGame Over')
```

```python
from IPython.display import HTML

HTML("""
<canvas id="game" width="800" height="400" style="border:1px solid black"></canvas>
<button onclick="spawnWave()">Start Wave</button>
<script>
const canvas = document.getElementById("game");
const ctx = canvas.getContext("2d");

let towers = [];
let enemies = [];
let bullets = [];
let money = 100;
let lives = 10;

canvas.addEventListener("click", e => {
  const rect = canvas.getBoundingClientRect();
  const x = e.clientX - rect.left;
  const y = e.clientY - rect.top;

  if (money >= 50 && Math.abs(y-200) > 40) {
    towers.push({x,y,range:120,cd:0});
    money -= 50;
  }
});

function spawnWave(){
  for(let i=0;i<5;i++){
    enemies.push({x:-i*60,y:200,hp:30});
  }
}

function update(){
  enemies.forEach(e=>e.x+=1);

  towers.forEach(t=>{
    if(t.cd>0) t.cd--;
    enemies.forEach(e=>{
      const d=Math.hypot(t.x-e.x,t.y-e.y);
      if(d<120 && t.cd<=0){
        bullets.push({x:t.x,y:t.y,tx:e});
        t.cd=40;
      }
    });
  });

  bullets.forEach(b=>{
    const dx=b.tx.x-b.x;
    const dy=b.tx.y-b.y;
    const d=Math.hypot(dx,dy);
    b.x+=dx/d*4;
    b.y+=dy/d*4;

    if(d<5){
      b.tx.hp-=10;
      b.dead=true;
    }
  });

  bullets=bullets.filter(b=>!b.dead);
  enemies=enemies.filter(e=>{
    if(e.hp<=0){money+=10;return false;}
    if(e.x>800){lives--;return false;}
    return true;
  });
}

function draw(){
  ctx.clearRect(0,0,800,400);
  ctx.fillStyle="tan";
  ctx.fillRect(0,180,800,40);

  towers.forEach(t=>{
    ctx.fillStyle="blue";
    ctx.beginPath();
    ctx.arc(t.x,t.y,10,0,Math.PI*2);
    ctx.fill();
  });

  enemies.forEach(e=>{
    ctx.fillStyle="red";
    ctx.beginPath();
    ctx.arc(e.x,e.y,10,0,Math.PI*2);
    ctx.fill();
  });

  bullets.forEach(b=>{
    ctx.fillStyle="black";
    ctx.fillRect(b.x,b.y,4,4);
  });

  ctx.fillStyle="black";
  ctx.fillText("Money: "+money,10,15);
  ctx.fillText("Lives: "+lives,10,30);
}

function loop(){
  update();
  draw();
  requestAnimationFrame(loop);
}

loop();
</script>
""")

```
